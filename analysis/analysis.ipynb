## Importing and cleaning the csv file

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
car_data = Path("../car details v4.csv")
```

```python
car_df = pd.read_csv(car_data)
car_df.head()
```

```python
car_df.count()
```

```python
car_df = car_df.dropna(how='any')
car_df.head()
```

```python
car_df.count()
```

```python
car_df['Location'].unique()
```

## Clean DataFrame - Includes Mumbai Data Only

```python
#new DF with only Mumbai data
clean_car = car_df.loc[car_df['Location'] == 'Mumbai']
clean_car.head()
```

## Make Analysis (Omar)


```python
#Group the data by car make and count the number of sales for each make in Mumbai
Make = clean_car['Make'].value_counts()
```

```python
Make_df = pd.DataFrame(Make)
Make_df
```

```python
#Plot the bar chart of car make

plt.figure(figsize=(10,6)) 
Make.plot(kind='bar')
plt.title('Vehicles Sold by Make')
plt.xlabel('Make')
plt.ylabel('Number of Sales')
plt.tight_layout()
plt.savefig("../analysis/MakeBar.png", dpi=150)
```

## Color Analysis (Cassie)
- Color: The top selling vehicles are white, silver, grey, blue, and red as represented by the bar graph. The pie chart is an additional chart to better visualize the breakdown by color as the bar chart only visualizes in one color (blue). Based on the analysis, LOCA Auto Merchant should consider selling white, silver, grey, blue, and red vehicles on the lot. 

```python
color_series = clean_car["Color"].value_counts()
```

```python
color_df = pd.DataFrame(color_series)
color_df
```

### Vehicles Sold by Color - Pie Chart

```python
labels = ["White", "Silver", "Grey", "Blue", "Red", "Black", "Brown", " ", " ", " ", " ", " ", " ", " ", " "]

count = [90, 49, 41, 38, 29, 18, 15, 6, 5, 4, 2, 2, 1, 1, 1]

colors = ["white", "silver", "grey", "blue", "red", "black", "brown", "chocolate", "gold", "maroon", "orange", "purple", "beige", "green", "yellow"]

plt.pie(count, labels=labels, colors=colors, shadow=True, startangle=135)
plt.title("Vehicles Sold by Color")
plt.tight_layout()
plt.savefig("../analysis/ColorPie2.png", dpi=200)
```

### Vehicles Sold by Color - Bar Chart

```python
color_chart = color_df.plot(kind="bar", facecolor='blue', figsize=(10,4),
                        title = "Vehicles Sold by Color",
                        xlabel="Color",
                        ylabel="Number of Sales")
xticklabels = color_df.index
color_chart.set_xticklabels(xticklabels, rotation=45)
plt.tight_layout()
plt.savefig("../analysis/ColorBar.png", dpi=150)
```

## Kilometer Analysis (Cassie)
 - Kilometers: The majority of vehicles sold have less than 60,000 kilometers. LOCA Auto Merchant should aim to sell vehicles that have lower kilometers.


### Vehicles Sold by Kilometer - Bar Chart

```python
bins=[0, 20000, 40000, 60000, 80000, 100000, 120000]
group_names = ["0-20,000", "20,000-40,000", "40,000-60,000", "60,000-80,000", "80,000-100,000", "100,000+"]

clean_car["Range"]=pd.cut(clean_car["Kilometer"], bins, labels=group_names)
clean_car.head()
```

```python
clean_car2 = clean_car.groupby(["Range"])["Make"].count()
clean_car2
```

```python
km_df = pd.DataFrame(clean_car2)
km_df
```

```python
km_bar = km_df.plot(kind="bar", facecolor='blue', figsize=(10,4),
                        title = "Vehicles Sold by Kilometers",
                        xlabel="Kilometer Range",
                        ylabel="Number of Sales")
plt.ylim(0, 100, 20)
xticklabels=["0-20,000", "20,000-40,000", "40,000-60,000", "60,000-80,000", "80,000-100,000", "100,000+"]
km_bar.set_xticklabels(xticklabels, rotation=45)
plt.tight_layout()
plt.savefig("../analysis/KilomBar.png", dpi=200)
```

## Number of Owners Analysis (Lauren)
- the majority of sales in this market are for autos with one previous owner

### Vehicles Sold by Number of Owners - Bar Chart
