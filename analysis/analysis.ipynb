## Clean DataFrame - Includes Mumbai Data Only

```python
#new DF with only Mumbai data
clean_car = car_df.loc[car_df['Location'] == 'Mumbai']
clean_car.head()
```

## Make Analysis (Omar)

- The car industry is fiercely competitive, with various factors influencing consumer preferences and purchasing decisions. One significant factor is the make or brand of a car, which plays a crucial role in determining sales figures. This analysis aims to explore the relationship between car make and sales, examining how brand reputation affects consumer behavior.
 
 - The brand name is a critical factor that impacts car sales. Established brands with a reliable track record and positive customer experiences enjoy a competitive advantage. Well-established brands are perceived as more trustworthy, leading consumers to choose them over lesser-known or emerging brands. For instance, brands like Toyota, Honda, and BMW have built reputations for creating high-quality vehicles, resulting in consistently high sales.
 
 - The influence of car make on sales figures should not be underestimated. When deciding to buy a car, consumers consider various important factors, including price, color, number of previous owners, make, and the number of kilometers on the car's odometer.

 - The highest volume selling makes are Hyundai and Maruti Suzuki.

### Vehicles Sold by Make - Bar Chart

```python
#Group the data by car make and count the number of sales for each make in Mumbai
Make = clean_car['Make'].value_counts()
```

```python
Make_df = pd.DataFrame(Make)
Make_df
```

```python
#Plot the bar chart of car make

plt.figure(figsize=(10,6)) 
Make.plot(kind='bar')
plt.title('Vehicles Sold by Make')
plt.xlabel('Make')
plt.ylabel('Number of Sales')
plt.tight_layout()
plt.savefig("../analysis/MakeBar.png", dpi=150)
```

## Kilometer Analysis (Cassie)

### Vehicles Sold by Kilometer - Bar Chart

```python
bins=[0, 20000, 40000, 60000, 80000, 100000, 120000]
group_names = ["0-20,000", "20,000-40,000", "40,000-60,000", "60,000-80,000", "80,000-100,000", "100,000+"]

clean_car["Range"]=pd.cut(clean_car["Kilometer"], bins, labels=group_names)
clean_car.head()
```

## Vehicles Sold by Price (Alex)
 - The lower the price of the vehicle, the higher the volume of sales. About 40% of sales were for vehicles below 650,000 rupee.

### Vehicles Sold by Price - Pie Chart

```python
bucket_size = 100000
min_price = clean_car['Price'].min()
max_price = clean_car['Price'].max()
clean_car['Price Buckets'] = pd.cut(clean_car['Price'], bins=np.arange(min_price, max_price + bucket_size, bucket_size))
#bucket
bucket_counts = clean_car['Price Buckets'].value_counts().sort_values(ascending=False)
top_10_buckets = bucket_counts[:12]
other_count = bucket_counts[12:].sum()
updated_bucket_counts = pd.concat([top_10_buckets, pd.Series(other_count, index=['Other'])])
```

```python

#Plot
plt.pie(updated_bucket_counts, labels=updated_bucket_counts.index, autopct='%1.1f%%')
plt.title('Vehicles Sold by Price')
plt.axis('equal')
plt.savefig("../analysis/PricePie.png", dpi=200)
plt.show()
```

### Vehicles Sold by Price - Scatter Plot

```python
#Scatter
bucket_size = 100000
# Create price buckets
clean_car['Price Buckets'] = pd.cut(clean_car['Price'], bins=np.arange(clean_car['Price'].min(), clean_car['Price'].max() + bucket_size, bucket_size))
# Convert the 'Price Buckets' column to a string representation
clean_car['Price Buckets'] = clean_car['Price Buckets'].astype(str)
# Count the number of cars in each price bucket
bucket_counts = clean_car['Price Buckets'].value_counts()
```

```python
# Create scatter plot
plt.scatter(bucket_counts.index, bucket_counts.values)
plt.title('Scatter Plot: Number of Cars Sold at Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Cars Sold')
plt.xticks(rotation=45)
plt.xticks(np.arange(0, len(bucket_counts), 10), rotation=45)
plt.savefig("../analysis/PriceScatter.png", dpi=200)
plt.show()
```
