```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
car_data = Path("../car details v4.csv")
```

```python
car_df = pd.read_csv(car_data)
car_df.head()
```

```python
car_df.count()
```

```python
car_df = car_df.dropna(how='any')
car_df.head()
```

```python
car_df.count()
```

```python
car_df['Location'].unique()
```

```python
#new DF with only Mumbai data
clean_car = car_df.loc[car_df['Location'] == 'Mumbai']
clean_car.head()
```

```python
clean_car.count()
```

```python
clean_car.columns
```

```python
color_series = clean_car["Color"].value_counts()
```

```python
color_df = pd.DataFrame(color_series)
color_df
```

```python
labels = ["White", "Silver", "Grey", "Blue", "Red", "Black", "Brown", " ", " ", " ", " ", " ", " ", " ", " "]

count = [90, 49, 41, 38, 29, 18, 15, 6, 5, 4, 2, 2, 1, 1, 1]

colors = ["white", "silver", "grey", "blue", "red", "black", "brown", "chocolate", "gold", "maroon", "orange", "purple", "beige", "green", "yellow"]

plt.pie(count, labels=labels, colors=colors, shadow=True, startangle=135)
plt.title("Vehicles Sold by Color")
plt.tight_layout()
plt.savefig("../analysis/ColorPie2.png", dpi=200)
```

```python
color_chart = color_df.plot(kind="bar", facecolor='orange', alpha=0.5, figsize=(10,4),
                        title = "Vehicles Sold by Color",
                        xlabel="Color",
                        ylabel="Number of Sales")
xticklabels = color_df.index
color_chart.set_xticklabels(xticklabels, rotation=45)
plt.tight_layout()
plt.savefig("../analysis/ColorBar.png", dpi=150)
```

```python
km_series = clean_car["Kilometer"].value_counts()
km_df = pd.DataFrame(km_series)
```

```python
#only want to keep the KM column - delete all others (class 4.3 has the binning example)
km_cols = clean_car["Kilometer"]
km_cols
```

```python
bins=[0, 20000, 40000, 60000, 80000, 100000, 120000]
group_names = ["0-20,000", "20,000-40,000", "40,000-60,000", "60,000-80,000", "80,000-100,000", "100,000+"]

clean_car["Range"]=pd.cut(clean_car["Kilometer"], bins, labels=group_names)
clean_car.head()
```

```python
clean_car2 = clean_car.groupby(["Range"])["Make"].count()
clean_car2
```

```python
km_df = pd.DataFrame(clean_car2)
km_df
```

```python
km_bar = km_df.plot(kind="bar", facecolor='pink', alpha=0.7, figsize=(10,4),
                        title = "Vehicles Sold by Kilometers",
                        xlabel="Kilometer Range",
                        ylabel="Number of Sales")
plt.ylim(0, 100, 20)
km_bar.set_xticklabels(xticklabels, rotation=45)
plt.tight_layout()
plt.savefig("../analysis/KilomBar.png", dpi=200)
```
