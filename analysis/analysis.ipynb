```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
car_data = Path("../car details v4.csv")
```

```python
car_df = pd.read_csv(car_data)
car_df.head()
```

```python
car_df.count()
```

```python
car_df = car_df.dropna(how='any')
car_df.head()
```

```python
car_df.count()
```

```python
car_df['Location'].unique()
```

```python
#new DF with only Mumbai data
clean_car = car_df.loc[car_df['Location'] == 'Mumbai']
clean_car.head()
```

## Lauren's analysis of autos sold vs previous number of owners
   - the majority of sales in this market of for autos with one previous owner
   - unregistered autos are fairly uncommon for sales as they are aged and need to be reregistered
   

```python
# bar chart
owner = clean_car['Owner'].value_counts()
owner_df = pd.DataFrame(owner)
plt.figure(figsize=(10,6))

owner_df.plot(kind='bar')
plt.title('Vehicles Sold by Multiple Owners')
plt.xlabel('Number of Previous owners per Auto Sales')
plt.ylabel('Number of Sales')

plt.tight_layout()
plt.savefig("../analysis/Owner.png", dpi=150)
```
