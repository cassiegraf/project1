```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
car_data = Path("../car details v4.csv")
```

```python
car_df = pd.read_csv(car_data)
car_df.head()
```

```python
car_df.count()
```

```python
car_df = car_df.dropna(how='any')
car_df.head()
```

```python
car_df.count()
```

```python
car_df['Location'].unique()
```

```python
#new DF with only Mumbai data
clean_car = car_df.loc[car_df['Location'] == 'Mumbai']
clean_car.head()
```

```python
#Group the data by car make and count the number of sales for each make in Mumbai

Make = clean_car['Make'].value_counts()
Make
```

```python
Make_df = pd.DataFrame(Make)
Make_df

```

```python
#Plot the bar chart of car make

plt.figure(figsize=(10,6)) 
Make.plot(kind='bar')
plt.title('Vehicles Sold by Make')
plt.xlabel('Make')
plt.ylabel('Number of Sales')
plt.tight_layout()
plt.savefig("../analysis/MakeBar.png", dpi=150)
```
